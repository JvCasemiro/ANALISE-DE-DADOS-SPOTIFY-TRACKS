```python
import pandas as pd
import numpy as np
```

```python
# Função usada para carregar os dados
def carregar_dados(caminho):
    df = pd.read_csv(caminho)
    return df
```

```python
# Função usada para mostrar quais os problemas dos dados
def checar_qualidade(df):
    print(f"Total de linhas e colunas inicial: {df.shape}")
    
    print("\nValores nulos por coluna:")
    print(df.isnull().sum())
    
    print("\nMúsicas duplicadas pelo ID:")
    print(df.duplicated(subset=['track_id']).sum())
    print("-" * 30)
```

```python
# Função para limpar e transformar
def limpar_dados(df):
    # Tirar a coluna 'Unnamed: 0', aparentemente o Kaggle exportou sem querer
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])

    # Remove a linha que tem artista e nome da música em branco
    df = df.dropna(subset=['artists', 'track_name'])

    # Remove a musica repetida para gêneros diferentes, mantendo só a primeira.
    df = df.drop_duplicates(subset=['track_id'], keep='first')

    # Deriva variável 1
    df['duration_min'] = round(df['duration_ms'] / 60000, 2)

    # Deriva variável 2 
    # Separa se a musica é de duração Padrão ou Longa e substitui em uma nova coluna
    df['categoria_duracao'] = np.where(df['duration_min'] > 5.0, 'Longa', 'Padrão')

    return df
```

```python
# Função para salvar
def salvar_dados(df, caminho):
    df.to_csv(caminho, index=False)
    print(f"Arquivo limpo salvo em: {caminho}")
```

```python
# Caminhos dos arquivos
arquivo_original = "../data/raw/dataset.csv"
arquivo_limpo = "../data/processed/dataset_clean.csv"

# Rodando as funções
dados = carregar_dados(arquivo_original)
checar_qualidade(dados)
dados_limpos = limpar_dados(dados)
salvar_dados(dados_limpos, arquivo_limpo)
```
