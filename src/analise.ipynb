```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados limpos
def carregar_dados_limpos(caminho):
    return pd.read_csv(caminho)
```

```python
# Pergunta 1: Quais os 10 gêneros musicais mais populares e menos populares em média?
def analisar_generos_populares(df):
    # Agrupa por gênero e calcula a média de popularidade dos 10 mais populares
    top_generos_high = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)
    
    # Criar o gráfico de barras
    plt.figure(figsize=(10, 5))
    top_generos_high.plot(kind='bar', color='blue', edgecolor='black')
    plt.title('Gêneros Musicais Mais Populares')
    plt.xlabel('Gênero')
    plt.ylabel('Popularidade')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def analisar_generos_menos_populares(df):
    # Agrupa por gênero e calcula a média de popularidade dos 10 menos populares
    piores_generos = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=True).head(10)
    
    plt.figure(figsize=(10, 5))
    piores_generos.plot(kind='bar', color='red', edgecolor='black')
    plt.title('Os 10 Gêneros Musicais Menos Populares')
    plt.xlabel('Gênero')
    plt.ylabel('Popularidade')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python

# Pergunta 2: Músicas explícitas(Com palavrões) são mais populares que as normais?
def analisar_explicitas(df):
    # Calcula a média de popularidade para explicita (True) e não explicita (False)
    media_explicita = df.groupby('explicit')['popularity'].mean()
    
    # Criar o gráfico
    plt.figure(figsize=(6, 5))
    # Trocando False para "Não Explicita" e True para "Explicita"
    media_explicita.index = ['Não Explícita', 'Explícita']
    
    media_explicita.plot(kind='bar', color=['green', 'orange'], edgecolor='black')
    plt.title('Músicas Explícitas vs Não Explícitas')
    plt.ylabel('Popularidade')
    plt.xticks(rotation=0)
    plt.show()
```

```python

# Pergunta 3: Qual é a distribuição do tempo de duração das músicas?
def analisar_duracao(df):
    # Separa apenas músicas com menos de 10 minutos para isolar podcasts.
    musicas_comuns = df[df['duration_min'] < 10]
    
    # Cria o grafico
    plt.figure(figsize=(8, 5))
    plt.hist(musicas_comuns['duration_min'], bins=30, color='purple', edgecolor='black')
    plt.title('Duração das Músicas por minuto')
    plt.xlabel('Duração (minutos)')
    plt.ylabel('Quantidade de Músicas')
    plt.show()
```

```python

# Carregando os dados limpos
caminho_limpo = r"../data/processed/dataset_clean.csv"    
df = carregar_dados_limpos(caminho_limpo)

#Rodando o grafico de generos mais populares    
analisar_generos_populares(df)

#Rodando o grafico de generos menos populares
analisar_generos_menos_populares(df)

#Rodando o grafico sobre musicas explicitas    
analisar_explicitas(df)

#Rodando o grafico sobre duração das musicas
analisar_duracao(df)
```
